```python
##### from skimage import io
from skimage import draw
img = io.imread('puppy.jpg')
print(img.shape)
def rectangle(x, y, w, h):
    rr, cc = [x, x + w, x + w, x], [y, y, y + h, y + h]
    return (draw.polygon(rr, cc))
rr, cc = rectangle(10, 10, 50, 50)
img[rr, cc] = 1
io.imshow(img)    
```
