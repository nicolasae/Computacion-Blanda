```python
from skimage import io
from skimage import draw
from skimage.draw import bezier_curve
img = io.imread('puppy.jpg')
print(img.shape)
# Define las coordenadas de la curva Bezier
x, y = draw.bezier_curve(0, 0, 50, 100, 100, 30, 2)
# Dibuja la curva Bezier
img[x, y] = 1
# Muestra la imagen
io.imshow(img)
```
