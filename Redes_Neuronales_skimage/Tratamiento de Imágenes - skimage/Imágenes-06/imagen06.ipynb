```python
# Importar librerías
from skimage import io
from skimage import color
from pylab import *
import numpy as np
import pandas as pd
# Leer imagen
img = io.imread('puppy.jpg')
# Convertir a YUV
img_yuv = color.rgb2yuv(img)
# Convertir de vuelta a RGB
img_rgb = color.yuv2rgb(img_yuv)

img_yuv = img_yuv/np.amax(img_yuv)
img_yuv = np.clip(img_yuv, 0, 1)

img_rgb = img_rgb/np.amax(img_rgb)
img_rgb = np.clip(img_rgb, 0, 1)

# Mostrar ambas figuras
figure(0)
io.imshow(img_yuv)
figure(1)
io.imshow(img_rgb)
```
