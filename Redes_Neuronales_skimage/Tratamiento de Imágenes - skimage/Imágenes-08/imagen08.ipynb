```python
# Importar librerías
from skimage import io
from skimage import color
from pylab import *
import numpy as np
import pandas as pd
# Leer imagen
img = io.imread('puppy.jpg')
# Convertir a YUV
img_ypbpr= color.rgb2ypbpr(img)
# Convertir atrás a RGB
img_rgb= color.ypbpr2rgb(img_ypbpr)

img_ypbpr = img_ypbpr/np.amax(img_ypbpr)
img_ypbpr = np.clip(img_ypbpr, 0, 1)

img_rgb = img_rgb/np.amax(img_rgb)
img_rgb = np.clip(img_rgb, 0, 1)

# Mostrar ambas figuras
figure(0)
io.imshow(img_ypbpr)
figure(1)
io.imshow(img_rgb)
```
