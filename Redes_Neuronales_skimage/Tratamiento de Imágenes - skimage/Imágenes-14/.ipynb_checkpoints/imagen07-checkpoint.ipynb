```python
#Import libraries
from skimage import io
from skimage import color
from pylab import *
import numpy as np
import pandas as pd
#Read image
img = io.imread('puppy.jpg')
#Convert to YUV
img_yiq = color.rgb2yiq(img)
#Convert back to RGB
img_rgb = color.yuv2rgb(img_yiq)

img_yiq = img_yiq/np.amax(img_yiq)
img_yiq = np.clip(img_yiq, 0, 1)

img_rgb = img_rgb/np.amax(img_rgb)
img_rgb = np.clip(img_rgb, 0, 1)

#Show both figures
figure(0)
io.imshow(img_yuv)
figure(1)
io.imshow(img_rgb)
```
