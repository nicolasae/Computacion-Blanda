```python
from skimage import io
from skimage import color
from skimage import data
from pylab import *
import numpy as np
import pandas as pd
img = io.imread('puppy.jpg')
# Convertir a XYZ
img_xyz = color.rgb2xyz(img)
# Convertir de vuelta a RGB
img_rgb = color.xyz2rgb(img_xyz)

# Si la intensidad del color es de punto flotante, entonces matplotlib espera que 
# los valores estn entre 0 y 1. Si la intensidad de color es de tipo entero, entonces 
# matplotlib espera que los valores estn en el rango entre 0 a 255. Por lo tanto, 
# una solucin consiste en escalar todos los nmeros dividindolos entre el valor
# mximo, de modo que el rango est entre 0 y 1 (para ello se utiliza clip). 
# Esto se puede conseguir con las # siguientes instruciones: 
# (en caso contrario, se genera un error)

img_xyz = img_xyz/np.amax(img_xyz)
img_xyz = np.clip(img_xyz, 0, 1)

# Muestra ambas figutas
figure(0)
io.imshow(img_xyz)
figure(1)
io.imshow(img_rgb)
```
